```python
# Read CSV for data

path= "/content/drive/MyDrive/Movie Analysis/movies.csv"

df = pd.read_csv(path)
```

```python
#View dataset

df.head()
```

```python
# Checking for missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
 #Remove null values
 
 df = df.dropna()
```

```python
# Checking for removed missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
df.dtypes
```

```python
#Change budget and gross columns data type to remove decimal

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')
```

```python
df.head()
```

```python
#Create corrected year column
df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
```

```python
df.head()
```

```python
# Sort gross profit in descending order

df.sort_values(by=['gross'], inplace=False, ascending=False)

df.head()
```

```python
pd.set_option('display.max_rows', None)
```

```python
# Check and drop any duplicates

df['company'].drop_duplicates().sort_values(ascending=False)

df.head()
```

```python
df.head()
```

```python
#Checking budget and gross correlation with a scatter plot

plt.scatter(x=df['budget'], y=df['gross'], c='cornflowerblue', marker='*')

plt.title('Movie Budget vs Gross Earnings')

plt.xlabel('Budget')
plt.ylabel('Gross Earnings')

plt.show()
```

```python
df.head()
```

```python
# Plot budget vs gross using seaborn

sns.regplot(x='budget', y='gross', data=df, marker='*', scatter_kws={"color": "cornflowerblue"}, line_kws={"color": "black"})
```

```python
#Checking correlation

df.corr()
```

```python
correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Numeric Movie Features Correlation Matrix')
plt.xlabel('Movie Features (Numeric)')
plt.ylabel('Movie Features (Numeric)')

plt.show
```

```python
#Explore company

df.head()
```

```python
df_numeric = df

for col_name in df_numeric.columns:
    if(df_numeric[col_name].dtype =='object'):
        df_numeric[col_name] = df_numeric[col_name].astype('category')
        df_numeric[col_name] = df_numeric[col_name].cat.codes
        
df_numeric.head()
```

```python
correlation_matrix = df_numeric.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Numeric Movie Features Correlation Matrix')
plt.xlabel('Movie Features (Numeric)')
plt.ylabel('Movie Features (Numeric)')

plt.show
```

```python
df_numeric.corr()
```

```python
corr_matrix = df_numeric.corr()

corr_pairs = corr_matrix.unstack()

corr_pairs.head()
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs.head()
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
